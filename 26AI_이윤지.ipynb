```python
data_1 = data[['경찰서', '2020년', '2021년', '2022년', '2023년', '2024년']]
```

```python
data_1.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl

plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.figure(figsize=(12,6))

plt.bar(data_1['경찰서'], data_1['2024년'], color="#b060c7")
plt.xticks(rotation=45)

plt.title("2024년 경찰서별 경찰관 수")
plt.xlabel("경찰서")
plt.ylabel("경찰관 수")

plt.show()
```

```python
# '경찰서'를 index로 설정
data_1 = data_1.set_index('경찰서')

# 연도별 데이터 추출
years = ['2020년', '2021년', '2022년', '2023년', '2024년']
df_years = data_1[years]

df_years.T.plot(figsize=(14, 8), marker='o', alpha=0.5, legend=False)
plt.title('경찰서별 2020~2024년 경찰관 수 변화')
plt.xlabel('연도')
plt.ylabel('경찰관 수')
plt.grid(True)
plt.tight_layout()
plt.show()
```
